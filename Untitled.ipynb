```python
import numpy as np
import seaborn as sns
```

```python
def moving_avg(a):
    """Carry out a moving average over array a, without changing the end values.
    """
    b = []
    for i in range(1,len(a)-1):
        b.append((a[i-1]+8*a[i]+a[i+1])/10)
    return [a[0]] + b + [a[-1]]
```

```python
boundary = 1.
Ncell = 50
Nt = 50
u_hist = []
u = [boundary] + [-0.99/(Ncell+1) * i + 0.99 for i in range(1, Ncell)]
u_hist.append(u)
for n in range(Nt):
    u_new = [0]*Ncell
    for i in range(len(u)):
        if i==0:
            u_new[i] = boundary
        elif i==Ncell-1:
            u_new[i] = u[i] - 0.05*(u[i-1] - u[i])
        else:
            u_new[i] = u[i] - 0.05*(u[i-1] - 2*u[i] + u[i+1])

    #u_new = moving_avg(u_new)
    u_hist.append(u_new)
    u = u_new

```

```python
sns.heatmap(u_hist)
```

```python
sns.lineplot(u_hist[0])
sns.lineplot(u_hist[int(Nt/2)])
sns.lineplot(u_hist[-1])
```

```python
boundary = 1.
Ncell = 50
Nt = 50
u = u_hist[-1]
u_hist = []
u_hist.append(u)
for n in range(Nt):
    u_new = [0]*Ncell
    for i in range(len(u)):
        if i==0:
            u_new[i] = boundary
        elif i==Ncell-1:
            u_new[i] = u[i] + 0.05*(u[i-1] - u[i])
        else:
            u_new[i] = u[i] + 0.05*(u[i-1] - 2*u[i] + u[i+1])
    u_hist.append(u_new)
    u = u_new
```

```python
sns.heatmap(u_hist)
```

```python
sns.lineplot(u_hist[0])
sns.lineplot(u_hist[int(Nt/2)])
sns.lineplot(u_hist[-1])
```

Forward heat equation:
each mode term has a time suppression like exp(-k n^2 t)
small modes (large n) will decay fast, as if n is large, -n^2 t is large and -ve, suppressing the exponent.
large modes (small n) decay slowest.

Backward equation:
each mode term has a time boost like exp(k n^2 t)
small modes (large n) are boosted quickly, as they have large exponents.
