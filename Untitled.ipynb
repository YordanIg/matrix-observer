```python
def numerical_log_derivative(x, fx):
    """
    Numerically calculates the log-derivative d(log f) / d(log x).

    Args:
        x (np.ndarray): The independent variable array.
        fx (np.ndarray): The function values f(x) corresponding to x.

    Returns:
        np.ndarray: The numerical log-derivative.
    """
    # Calculate df/dx using numpy.gradient
    df_dx = np.gradient(fx, x)

    # Calculate the log-derivative
    # d log f / d log x = (x / f) * (df / dx)
    log_deriv = (x / fx) * df_dx

    return log_deriv
```

```python
def take_log_deriv_of_sky_model(mod, nuarr):
    T_ = []
    for nu in nuarr:
        m = mod.generate(nu)
        NSIDE = hp.npix2nside(len(m))
        vec = hp.ang2vec(np.pi / 3, -np.pi * 2 / 4)
        ipix_disc = hp.query_disc(nside=NSIDE, vec=vec, radius=np.radians(20))
        T = np.mean(m[ipix_disc])
        m[ipix_disc] = m.max()
        T_.append(T)
    return numerical_log_derivative(nuarr, T_)
```

```python
def reconstruct_map_from_svd(a_ml, mat_Vh, truncation_idx):
    """
    For a given map estimate a_ml with the svd basis matrix mat_Vh, project the
    map estimate onto the 
    """
    a_trunc = np.zeros(len(a_ml))
    for i in range(truncation_idx):
        basis_vec = mat_Vh[i]
        a_trunc += np.dot(basis_vec, a_ml) * basis_vec
    return a_trunc
```

Forward heat equation:
each mode term has a time suppression like exp(-k n^2 t)
small modes (large n) will decay fast, as if n is large, -n^2 t is large and -ve, suppressing the exponent.
large modes (small n) decay slowest.

Backward equation:
each mode term has a time boost like exp(k n^2 t)
small modes (large n) are boosted quickly, as they have large exponents.
